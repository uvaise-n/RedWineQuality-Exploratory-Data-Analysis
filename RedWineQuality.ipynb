Correlation between variables


```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),cmap='Blues',annot=True) 
```


Observations: 

How to check correlation: 
if corr > 0.5 OR corr < -0.5 = strong/ very good correlation
if corr is between -0.1 to -0.5 or +0.1 and +0.5 = good correlation
if corr is between -0.1 and +0.1 = bad/ weak correlation
+corr = directly proportional ; -corr= inversely proportional

Fixed Acidity, citric acid, sulphates and alcohol has a good positive correlation. Alchohol infact has the strongestt correlation of the variables
Volatile acidity,  clorides, total sulphur dioxide, density has a good negative correaltion. 
It is also to be noted that fixed acidity and density has a stron correlation
Residual sugar and pH has almost no correlation to quality


Check for Outliers

```python
l = df.columns.values
number_of_columns=12
number_of_rows = len(l)-1/number_of_columns
plt.figure(figsize=(number_of_columns,5*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows + 1,number_of_columns,i+1)
    sns.set_style('whitegrid')
    sns.boxplot(df[l[i]],color='green',orient='v')
    plt.tight_layout()
```

There are outliers in all variables

Check for skewness

```python
plt.figure(figsize=(2*number_of_columns,5*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows + 1,number_of_columns,i+1)
    sns.distplot(df[l[i]],kde=True) 
```
